# COVID 19 en Colombia

## ¿Cuándo Colombia alcanzará 1.000.000 de casos positivos?

A continuación haremos una predicción en base a la información actual sobre el COVID 19 en Colombia.

```python
'''
Primero importamos las librerias necesarias para procesar la información
y hacer las respectiva proyección y predicción
'''
import os
import numpy as np
import scipy as sp
import bisect
from scipy.optimize import fsolve
import matplotlib.pyplot as plt
from utils import DATA_DIR, CHART_DIR

'''
Ahora procedemos a hacer la lectura de los datos y almacenarlos en una lista
'''
data = np.genfromtxt("covidco.csv", delimiter=',', dtype=int, usecols=5)[64:]

'''
Separamos los datos dependientes e independientes para empezar su procesamiento
'''
x = np.array([*range(1,214)])
y = np.array(data)

'''
Función para graficar los modelos
'''
colors = ['g', 'k', 'b', 'm', 'r']
linestyles = ['-', '-.', '--', ':', '-']
month_starts = [1, 26, 57, 87, 118, 149, 179, 210, 240, 271]
month_names = ['Marzo','Abril','Mayo','Junio','Julio',
               'Agosto','Septiembre','Octubre','Noviembre','Diciembre'] 

def plot_models(x, y, models, fname, mx=None, ymax=None, xmin=None):
    ''' dibujar datos de entrada '''
    
    plt.figure(num=None, figsize=(8, 6))
    plt.clf()
    plt.scatter(x, y, s=10)
    
    plt.xticks(month_starts, month_names, rotation='vertical')
    
    plt.title("COVID 19 COLOMBIA")
    plt.xlabel("Mes")
    plt.ylabel("Casos Positivos")

    if models:
        if mx is None:
            mx = np.linspace(0, x[-1], 1000)
        for model, style, color in zip(models, linestyles, colors):
            # print "Model:",model
            # print "Coeffs:",model.coeffs
            plt.plot(mx, model(mx), linestyle=style, linewidth=2, c=color)
            
        plt.legend(["d=%i" % m.order for m in models], loc="upper left")
        
    plt.autoscale(tight=True)
    plt.ylim(ymin=0)
    if ymax:
        plt.ylim(ymax=ymax)
    if xmin:
        plt.xlim(xmin=xmin)
    plt.grid(True, linestyle='-', color='0.75')
    plt.savefig(fname)
    
'''
Grafiquemos los datos que tenemos
'''
plot_models(x, y, None, os.path.join(CHART_DIR, "COVID19_01.png"))

'''
Creamos los modelos de datos para diferentes grados de ecuación
'''
f1 = sp.poly1d(np.polyfit(x, y, 1))
f2 = sp.poly1d(np.polyfit(x, y, 2))
f3 = sp.poly1d(np.polyfit(x, y, 3))
f10 = sp.poly1d(np.polyfit(x, y, 10))

'''
Graficamos los modelos de datos
'''
plot_models(x, y, [f1], os.path.join(CHART_DIR, "COVID19_02.png"))
plot_models(x, y, [f1, f2], os.path.join(CHART_DIR, "COVID19_03.png"))
plot_models(x, y, [f1, f2, f3, f10], os.path.join(CHART_DIR,"COVID19_04.png"))

'''
Como no hay puntos de inflexión aparentes, procederemos a realizar el entrenamiento
y la predicción respectiva
'''

'''
Función de error
'''
def error(f, x, y):
    return np.sum((f(x) - y) ** 2)

'''
Extrapolamos para que se proyecten respuestas en el futuro
'''
plot_models(x, y, [f1, f2, f3, f10], os.path.join(CHART_DIR, "COVID19_05.png"),
            mx=np.linspace(0, 271, 50), ymax=1000000, xmin=0)

'''
Separamos los datos de prueba y entrenamiento
'''

'''
Frac es la proporción que vamos a obtener de todos los datos.
En este caso es un 30%
'''
frac = 0.3
               
# Está es la cantidad de valores a tomar de la población de datos
split_idx = int(frac * len(x))
               
# Obtenemos los indices que vamos a extraer de manera aleatoria
shuffled = sp.random.permutation(list(range(len(x))))
               
'''
Tomamos los valores para el test de los indices
hasta alcanzar el 30% de la población
'''
test = sorted(shuffled[:split_idx])
               
'''
Tomamos los valores que sobran de la lista anterior,
es decir, el otro 70% de la población
'''
train = sorted(shuffled[split_idx:])

'''
Hacemos un entrenamiento con la muestra obtenida
para diferentes modelos de datos
'''
ft1 = sp.poly1d(np.polyfit(x[train], y[train], 1))
ft2 = sp.poly1d(np.polyfit(x[train], y[train], 2))
ft3 = sp.poly1d(np.polyfit(x[train], y[train], 3))
ft10 = sp.poly1d(np.polyfit(x[train], y[train], 10))
               
# Graficamos los modelos de datos que contienen respuestas futuras
plot_models(x, y, [ft1, ft2, ft3, ft10], os.path.join(CHART_DIR, "COVID19_06.png"), 
            mx=np.linspace(0, 271, 100), ymax=1000000, xmin=0)

'''
Verificamos el error de cada uno de los modelos con la muestra obtenida
para el test
'''
print("Prueba de error")
for f in [ft1, ft2, ft3, ft10]:
    print(f"Error d={f.order}: {error(f, x[test], y[test])}")

# Hacemos la predicción y la imprimimos
alcanzado_max = fsolve(ft2 - 1000000, x0=222)
interval = bisect.bisect_left(month_starts, alcanzado_max[0])
result = month_starts[interval] - alcanzado_max[0]
print(f'\nSe alcanza 1.000.000 de casos positivos el día {int(result)} de {month_names[interval-1]}')
```
